## Abstract
- 通过表格存储值函数会占用大量内存，此时自然的想法是将离散化的值参数化，从而使其有泛化的能力，求出未知state的值。
- 估计值函数是一个supervised learning, label可以使用MC-target或者TD-target, 后者用到了bootstrapping, 即使用estimator估计的值继续估计，所以他是semi-gradient descent的。
- RL的值函数估计需要注意non-stationary (policy changes, bootstrapping) 和 non-iid (correlated in time) data的问题
- 在control的过程中，可以证明线性的模型总是可以收敛的。其余的形式很那保证收敛，但却有用。
- Value-Based的方法核心是估计值函数，然后使用greedy policy improvement；也可以直接对policy进行估计。
- 策略梯度方法的核心原理是Policy Gradient Therom, $\nabla J(\boldsymbol{\theta}) \propto \sum_{s} \mu(s) \sum_{a} q_{\pi}(s, a) \nabla_{\boldsymbol{\theta}} \pi(a | s, \boldsymbol{\theta})$
- 通过添加BaseLine、Critic(值函数指导)等可以衍生出REINFORCE,TD-PD,AC,A3C,DDPG,PPO等算法。

## Value Function Approximaton
当state过多时，没有足够的内存再使用tabular的形式存储离散的Value Function, 这时可以使用机器学习中的监督学习方法FA Function Approximation学习出$v(s, w)$, FA的过程首先要每个样本$S_t$对应的label的形式。通过前面的章节可以知道value function值即FA用到的label可以由以下方式得到：
- MC : $G_t$
- TD : $R_{t+1} + \gamma V(S_{t+1})$
- TD-nstep : $G_{t:t+n}$
- DP : $\mathbb{E}[R_{t+1} + \gamma V(s_{t+1})]$

## The prediction Objective
label定义之后，就需要确定Cost Function的形式。
$$
\overline{\mathrm{VE}}(\mathbf{w}) \doteq \sum_{s \in \mathcal{S}} \mu(s)\left[v_{\pi}(s)-\hat{v}(s, \mathbf{w})\right]^{2}
$$
其中, $\mu(s)$是state的分布情况, 在连续tasks中他是稳定的。

## Stochastic-gradient and Semi-gradient Methods
### SGD
有了label和损失函数，常规的优化方法就是梯度下降。 在RL问题中倾向于选择SGD：
$$
\begin{aligned} \mathbf{w}_{t+1} & \doteq \mathbf{w}_{t}-\frac{1}{2} \alpha \nabla\left[v_{\pi}\left(S_{t}\right)-\hat{v}\left(S_{t}, \mathbf{w}_{t}\right)\right]^{2} \\ &=\mathbf{w}_{t}+\alpha\left[v_{\pi}\left(S_{t}\right)-\hat{v}\left(S_{t}, \mathbf{w}_{t}\right)\right] \nabla \hat{v}\left(S_{t}, \mathbf{w}_{t}\right) \end{aligned}
$$
其中，梯度的计算方式为：
$$
\nabla f(\mathbf{w}) \doteq\left(\frac{\partial f(\mathbf{w})}{\partial w_{1}}, \frac{\partial f(\mathbf{w})}{\partial w_{2}}, \dots, \frac{\partial f(\mathbf{w})}{\partial w_{d}}\right)^{\top}
$$
为什么使用SGD而不是更精准的批梯度下降？
> Remember that we do not seek or expect to find a value function that has zero error for all states, but only an approximation that balances the errors in different states. If we completely corrected each example in one step, then we would not find such a balance

因为每个state的对应值是动态变化的，每次梯度只下降一小步有利于我们收敛到每个state的最小值。
### semi-GD
前面的梯度下降公式还不能用于计算，将$V(S_{t+1}, \mathbf{w})$更换为估计（label）后：
$$
\mathbf{w}_{t+1} \doteq \mathbf{w}_{t}+\alpha\left[U_{t}-\hat{v}\left(S_{t}, \mathbf{w}_{t}\right)\right] \nabla \hat{v}\left(S_{t}, \mathbf{w}_{t}\right)
$$
$U_t$可以是无偏估计如MC和DP的形式。 如果采用如TD的bootstrapping的形式，此时的方法就称为semi-GD的。
> They take into account the effect of changing the weight vector $w_t$ on the estimate, but ignore its effect on the target. They include only a part of the gradient and, accordingly, we call them semi-gradient methods.

之前已经做过TD和MC等无偏估计的比较，事实证明bootstrapp的方法依然能收敛到局部最优且速度快。

## Linear  Model
在FA问题中最经典的模型就是线性模型，同理，我们可以吧要估计的value function定义为线性的形式：
$$
\hat{v}(s, \mathbf{w}) \doteq \mathbf{w}^{\top} \mathbf{x}(s) \doteq \sum_{i=1}^{d} w_{i} x_{i}(s)
$$
线性的优势是，可以证明线性总是能收敛的。

**Proof the convergence of Linear Model:**
$$
\begin{aligned} \mathbf{w}_{t+1} & \doteq \mathbf{w}_{t}+\alpha\left(R_{t+1}+\gamma \mathbf{w}_{t}^{\top} \mathbf{x}_{t+1}-\mathbf{w}_{t}^{\top} \mathbf{x}_{t}\right) \mathbf{x}_{t} \\ &=\mathbf{w}_{t}+\alpha\left(R_{t+1} \mathbf{x}_{t}-\mathbf{x}_{t}\left(\mathbf{x}_{t}-\gamma \mathbf{x}_{t+1}\right)^{\top} \mathbf{w}_{t}\right) \end{aligned}
$$
规划为矩阵形式：
$$
\mathbb{E}\left[\mathbf{w}_{t+1} | \mathbf{w}_{t}\right]=\mathbf{w}_{t}+\alpha\left(\mathbf{b}-\mathbf{A} \mathbf{w}_{t}\right)
$$
$$
\mathbb{E}\left[\mathbf{w}_{t+1} | \mathbf{w}_{t}\right]=(\mathbf{I}-\alpha \mathbf{A}) \mathbf{w}_{t}+\alpha \mathbf{b}
$$
可以看出这是一个差分方程，由线性代数知差分方程的通项公式形式为:
$$
w_{t} = c_1\lambda_1^tx_1 + ... c_n\lambda_n^{t}x_n + \alpha b
$$
其中$\lambda$是矩阵$(I-\alpha A)$特征值，$x$是特征向量，$c$满足$Xc=x_0$， 那么只有当所有的$0 \le \lambda \lt 1$时，前一项收敛到0，而$b$是常量。

$$
\begin{aligned}
eigenValue(I-\alpha A) & = 1 - eigenValue(\alpha A) \\
&= 1 - \alpha * eigenValue(A)
\end{aligned}
$$
因为$\alpha$是在$(0,1]$的，所以只要$A$的所有特征值大于0即可，那么A必须是一个**正定矩阵**。

证明问题转变为证明矩阵$A$是正定矩阵：

$$
\begin{aligned} \mathbf{A} &=\sum_{s} \mu(s) \sum_{a} \pi(a | s) \sum_{r, s^{\prime}} p\left(r, s^{\prime} | s, a\right) \mathbf{x}(s)\left(\mathbf{x}(s)-\gamma \mathbf{x}\left(s^{\prime}\right)\right)^{\top} \\ &=\sum_{s} \mu(s) \sum_{a} p\left(s^{\prime} | s\right) \mathbf{x}(s)\left(\mathbf{x}(s)-\gamma \mathbf{x}\left(s^{\prime}\right)\right)^{\top} \\ &=\sum_{s} \mu(s) \mathbf{x}(s)\left(\mathbf{x}(s)-\gamma \sum_{s^{\prime}} p\left(s^{\prime} | s\right) \mathbf{x}\left(s^{\prime}\right)\right)^{\top} \\ &=\mathbf{X}^{\top} \mathbf{D}(\mathbf{I}-\gamma \mathbf{P}) \mathbf{X} \end{aligned}
$$

再转化为证明矩阵$D(I - \gamma P))$为正定矩阵，$D$是state分布的对角矩阵，$P$是对称矩阵，**同时也是Markov 矩阵**。所以$P$的特征值满足：
- 一个特征值为1
- 其余特征值小于1

所以$D(I-\gamma P)$的特征值都满足大于0, 是一个正定矩阵。

综上证明了Linear Model一定会收敛的原因。

## FA Example Mountain-Car
Mountain-Car是不断地调整小车的动作使其到达山顶的环境。可以定义成离散或连续的。

```python
import gym
import itertools
import matplotlib.pyplot as plt
import numpy as np
import sklearn.pipeline
import sklearn.preprocessing
from sklearn.linear_model import SGDRegressor
from sklearn.kernel_approximation import RBFSampler
plt.style.use('ggplot')
```

```python
env = gym.envs.make("MountainCar-v0")
observation_examples = np.array([env.observation_space.sample() for x in range(10000)])

# 归一化
scaler = sklearn.preprocessing.StandardScaler()
scaler.fit(observation_examples)

featurizer = sklearn.pipeline.FeatureUnion([
        ("rbf1", RBFSampler(gamma=5.0, n_components=100)),
        ("rbf2", RBFSampler(gamma=2.0, n_components=100)),
        ("rbf3", RBFSampler(gamma=1.0, n_components=100)),
        ("rbf4", RBFSampler(gamma=0.5, n_components=100))
        ])
featurizer.fit(scaler.transform(observation_examples))
```

```python
# 定义approximator
class Estimator():
    def __init__(self):
        self.models = []
        for _ in range(env.action_space.n):
            model = SGDRegressor(learning_rate='constant')
            model.partial_fit([self.featurize_state(env.reset())], [0])
            self.models.append(model)

    def featurize_state(self, state):
        scaled = scaler.transform([state])
        feature = featurizer.transform(scaled)
        return feature[0]
    
    def predict(self, s, a=None):
        s = self.featurize_state(s)
        if a:
            return self.models[a].predict([s])
        else:
            return np.array([model.predict([s]) for model in self.models])
        
    def update(self, s, a, y):
        s = self.featurize_state(s)
        self.models[a].partial_fit([s], [y])
```

```python
def make_epsilon_greedy_policy(estimator, epsilon, nA):
    def policy_fn(obs):
        A = np.ones(nA, dtype='float') * epsilon / nA
        q_values = estimator.predict(obs)
        best_action = np.argmax(q_values)
        A[best_action] += (1.0 - epsilon)
        return A
    return policy_fn
```

```python
from util import plotting
from tqdm import tqdm
def q_learning_fa(env, estimator, num_episode, discount_factor=1.0, epsilon=0.1, epsilon_decay=1.0):
    stats = plotting.EpisodeStats(
                episode_lengths=np.zeros(num_episode),
                episode_rewards=np.zeros(num_episode),
                methods='q_learning_function_approximation')
    for i_episode in tqdm(range(num_episode)):
        policy = make_epsilon_greedy_policy(estimator, epsilon * epsilon_decay**i_episode, env.action_space.n)
        state = env.reset()
        for t in itertools.count():
            action_probs = policy(state)
            action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
            
            next_state, reward, done, _ = env.step(action)
            
            stats.episode_lengths[i_episode] = t
            stats.episode_rewards[i_episode] += reward
            
            q_value = estimator.predict(next_state)
            target = reward + discount_factor * np.max(q_value)
            estimator.update(state, action, target)
            
            if done:
                break
            state = next_state
    return stats
```

```python
# 测试
estimator = Estimator()
stats = q_learning_fa(env, estimator, 100, epsilon=0.0)
plotting.plot_cost_to_go_mountain_car(env, estimator)
plotting.plot_episode_stats(stats, smoothing_window=25)
```

## The Policy Gradient Theorem
在此之前我们都是估计值函数，再由值函数得出策略。现在我们考虑直接估计出参数化的策略$\pi(s,a|\theta)$

首先确定FA的prediction objective:
$$
J(\boldsymbol{\theta}) \doteq v_{\pi_{\theta}}\left(s_{0}\right)
$$

可以证明得到：
$$
\nabla J(\boldsymbol{\theta}) \propto \sum_{s} \mu(s) \sum_{a} q_{\pi}(s, a) \nabla_{\boldsymbol{\theta}} \pi(a | s, \boldsymbol{\theta})
$$
上式就是**policy gradient theorem**， 是所有PD算法的理论基础。

## REINFORCE
与值函数的FA方法一样，PD也有无偏估计的版本，即用MC的思想：
$$
\begin{aligned} \nabla J(\boldsymbol{\theta}) &=\mathbb{E}_{\pi}\left[\sum_{a} \pi\left(a | S_{t}, \boldsymbol{\theta}\right) q_{\pi}\left(S_{t}, a\right) \frac{\nabla_{\boldsymbol{\theta}} \pi\left(a | S_{t}, \boldsymbol{\theta}\right)}{\pi\left(a | S_{t}, \boldsymbol{\theta}\right)}\right] \\ &=\mathbb{E}_{\pi}\left[q_{\pi}\left(S_{t}, A_{t}\right) \frac{\nabla_{\boldsymbol{\theta}} \pi\left(A_{t} | S_{t}, \boldsymbol{\theta}\right)}{\pi\left(A_{t} | S_{t}, \boldsymbol{\theta}\right)}\right] \\ &=\mathbb{E}_{\pi}\left[G_{t} \frac{\nabla_{\boldsymbol{\theta}} \pi\left(A_{t} | S_{t}, \boldsymbol{\theta}\right)}{\pi\left(A_{t} | S_{t}, \boldsymbol{\theta}\right)}\right]
\end{aligned}
$$

从而得到参数的更新公式：
$$
\boldsymbol{\theta}_{t+1} \doteq \boldsymbol{\theta}_{t}+\alpha G_{t} \frac{\nabla_{\boldsymbol{\theta}} \pi\left(A_{t} | S_{t}, \boldsymbol{\theta}_{t}\right)}{\pi\left(A_{t} | S_{t}, \boldsymbol{\theta}_{t}\right)}
$$


## Reducing Variance Using a Critic
MC的无偏估计会带来较大方差，减小方差可以将引入Critic：
$$
\begin{aligned} \nabla_{\theta} J(\theta) & \approx \mathbb{E}_{\pi_{\theta}}\left[\nabla_{\theta} \log \pi_{\theta}(s, a) Q_{w}(s, a)\right] \\ \Delta \theta &=\alpha \nabla_{\theta} \log \pi_{\theta}(s, a) Q_{w}(s, a) \end{aligned}
$$

## Reducing Variance Using a Baseline
引入baseline后并不会改变原有的期望值：
$$
\begin{aligned} \mathbb{E}_{\pi_{\theta}}\left[\nabla_{\theta} \log \pi_{\theta}(s, a) B(s)\right] &=\sum_{s \in \mathcal{S}} d^{\pi_{\theta}}(s) \sum_{a} \nabla_{\theta} \pi_{\theta}(s, a) B(s) \\ &=\sum_{s \in S} d^{\pi_{\theta}} B(s) \nabla_{\theta} \sum_{a \in \mathcal{A}} \pi_{\theta}(s, a) \\ &=0 \end{aligned}
$$

baseline的选取只要跟action无关，即至于state有关即可，最直接的选择就是value function： $B(s) = V^{\pi_\theta}(s)$
然后我们可以写出带有**Advantage Function**的公式：
$$
\begin{aligned} A^{\pi_{\theta}}(s, a) &=Q^{\pi_{\theta}}(s, a)-V^{\pi_{\theta}}(s) \\ \nabla_{\theta} J(\theta) &=\mathbb{E}_{\pi_{\theta}}\left[\nabla_{\theta} \log \pi_{\theta}(s, a) A^{\pi \theta}(s, a)\right] \end{aligned}
$$

但是不需要估计两个函数，由：
$$
\delta^{\pi_{\theta}}=r+\gamma V^{\pi_{\theta}}\left(s^{\prime}\right)-V^{\pi_{\theta}}(s)
$$
$$
\begin{aligned} \mathbb{E}_{\pi_{\theta}}\left[\delta^{\pi_{\theta}} | s, a\right] &=\mathbb{E}_{\pi_{\theta}}\left[r+\gamma V^{\pi_{\theta}}\left(s^{\prime}\right) | s, a\right]-V^{\pi_{\theta}}(s) \\ &=Q^{\pi_{\theta}}(s, a)-V^{\pi_{\theta}}(s) \\ &=A^{\pi_{\theta}}(s, a) \end{aligned}
$$
可知 可以用TD error来当做Advantage function.
$$
\nabla_{\theta} J(\theta)=\mathbb{E}_{\pi_{\theta}}\left[\nabla_{\theta} \log \pi_{\theta}(s, a) \delta^{\pi_{\theta}}\right]
$$
$$
\delta_{v}=r+\gamma V_{v}\left(s^{\prime}\right)-V_{v}(s)
$$

##  Policy Parameterization for Continuous Actions
连续空间的action是连续的，需要对policy做参数化分布处理：
$$
\pi(a | s, \boldsymbol{\theta}) \doteq \frac{1}{\sigma(s, \boldsymbol{\theta}) \sqrt{2 \pi}} \exp \left(-\frac{(a-\mu(s, \boldsymbol{\theta}))^{2}}{2 \sigma(s, \boldsymbol{\theta})^{2}}\right)
$$
然后对$\mu(s,\theta)$和$\sigma(s, \theta)$进行估计。

## Example: Actor Critic
这里用一层神经网络实现一个AC的算法,环境使用连续状态的MounrainCar.

```python
N_A = env.action_space.shape[0]
A_BOUND = [env.action_space.low, env.action_space.high]
def featurize_state(state):
    """
    Returns the featurized representation for a state.
    """
    scaled = scaler.transform(state.reshape(1, -1))
    featurized = featurizer.transform(scaled)
    return featurized
```

```python
import tensorflow as tf
class PolicyEstimator():
    def __init__(self, learning_rate=0.001, scope='policy_estimator'):
        with tf.variable_scope(scope):
            self.state = tf.placeholder(tf.float32, [None, 400], name='state')
            self.target = tf.placeholder(tf.float32, [None, 1], name='target')
            self.action = tf.placeholder(tf.float32, name='action')
            
            w_init = tf.random_normal_initializer(0., .1)
            l_a = tf.layers.dense(self.state, 200, tf.nn.relu6, kernel_initializer=w_init, name='la')
            self.mu = tf.layers.dense(l_a, N_A, tf.nn.tanh, kernel_initializer=w_init, name='mu')
            self.sigma = tf.layers.dense(l_a, N_A, tf.nn.softplus, kernel_initializer=w_init, name='sigma')
            
            self.normal_dist = tf.contrib.distributions.Normal(self.mu, self.sigma)

            # Loss and train op
            a_prob = self.normal_dist.log_prob(self.action)
            self.loss = - a_prob * self.target
            # Add cross entropy cost to encourage exploration
            self.loss -= 1e-1 * self.normal_dist.entropy()
            
            self.chosen_action = self.normal_dist._sample_n(1)
            self.chosen_action = tf.clip_by_value(self.chosen_action, env.action_space.low[0], env.action_space.high[0])
            
            self.optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
            self.train_op = self.optimizer.minimize(
                self.loss, global_step=tf.contrib.framework.get_global_step())
            
    def predict(self, state, sess=None):
        sess = sess or tf.get_default_session()
        state = featurize_state(state)
        return sess.run(self.chosen_action, {self.state: state})
    
    def update(self, state, target, action, sess=None):
        sess = sess or tf.get_default_session()
        state = featurize_state(state)
        feed_dict = {
            self.state: state,
            self.action: action,
            self.target: target
        }
        _, loss = sess.run([self.train_op, self.loss], feed_dict)
        return loss
```

```python
class  ValueEstimator():
    def __init__(self, learning_rate=0.01, scope='value_estimator'):
        with tf.variable_scope(scope):
            w_init = tf.random_normal_initializer(0., .1)
            self.state = tf.placeholder(tf.float32, [None, 400], name='state')
            self.target = tf.placeholder(tf.float32, [None, 1], name='target')
            
            l_c = tf.layers.dense(self.state, 100, tf.nn.relu6, kernel_initializer=w_init)
            self.v = tf.layers.dense(l_c, 1, kernel_initializer=w_init, name='v')

            td = tf.subtract(self.target, self.v, name='TD_error')
            self.loss = tf.reduce_mean(tf.square(td))
            self.optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
            self.train_op = self.optimizer.minimize(self.loss, global_step=tf.contrib.framework.get_global_step())
            
    def predict(self, state, sess=None):
        sess = sess or tf.get_default_session()
        state = featurize_state(state)
        return sess.run(self.v, {self.state: state})
    
    def update(self, state, target, sess=None):
        sess = sess or tf.get_default_session()
        state = featurize_state(state)
        feed_dict = {
            self.state: state,
            self.target: target,
        }
        _, loss = sess.run([self.train_op, self.loss], feed_dict)
        return loss
```

```python
def actor_critic(env, policy_estimator, value_estimator, num_episodes=50, discount_factor=0.9):
    stats = plotting.EpisodeStats(
                episode_lengths=np.zeros(num_episodes),
                episode_rewards=np.zeros(num_episodes),
                methods='actor_critic')
    
    for i_episode in tqdm(range(num_episodes)):
        state = env.reset()
        for t in itertools.count():
            action = policy_estimator.predict(state)        
            next_state, reward, done, _ = env.step(action)
            
            stats.episode_lengths[i_episode] = t
            stats.episode_rewards[i_episode] += reward
            
            next_value = value_estimator.predict(next_state)
            td_target = reward + discount_factor * next_value
            td_error = td_target - value_estimator.predict(state)

            value_estimator.update(state, td_target)
            policy_estimator.update(state, td_error, action)

            if done:
                break
            state = next_state
    return stats
```

```python
import gym_gridworlds
env = gym.envs.make("MountainCarContinuous-v0")
tf.reset_default_graph()
value_estimator = ValueEstimator()
policy_estimator = PolicyEstimator()
global_step = tf.Variable(0, name="global_step", trainable=False)
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    stats = actor_critic(env, policy_estimator, value_estimator)
plotting.plot_episode_stats(stats)
```
