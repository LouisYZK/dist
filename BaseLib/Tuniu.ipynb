```python
# 文本预处理模块
import jieba
import codecs
import re
class preprocess():
    stopwords = []
    stopword_filepath="./stopwordList/stopword.txt"
    def __init__(self):
        self.__readin_stop()
    def __readin_stop(self):
        file_obj = codecs.open(self.stopword_filepath,'r','utf-8')
        while True:
            line = file_obj.readline()
            line=line.strip('\r\n')
            if not line:
                break
            self.stopwords.append(line)
        file_obj.close()
    def clean_doc(self,doc):
        # 汉字的Unicode范围为4e00-9fa5
        pattern = re.compile(r'[\u4e00-\u9fa5]+')
        filter_data = re.findall(pattern,doc)
        cleaned_doc = ''.join(filter_data)
        return cleaned_doc
    def cut(self,doc):
        seg = jieba.cut(doc)
        results = []
        for item in seg:
            if item in self.stopwords:
                continue
            results.append(item)
        return results
```

```python
# 产品供给特征结构图
from collections import Counter
# 特色用title列进行表示，目的地用overview列进行表示

feature = df['title'].values
desti = df['overview'].values
pre = preprocess()
# 尽管不愿用for循环，但是因为这一列的值类型不统一，不易处理异常
df['Feature'] = ''
df['Desti'] = ''
# pre.cut(pre.clean_doc(feature[0]))
F = []
for f in feature:
    F.append(pre.cut(pre.clean_doc(f)))
df['feature'] = F
# df.head()
```

```python
# 统计最高词频数模块
def flatten(items):
    for x in items:
        # 终止条件，检验是否为可迭代对象
        if hasattr(x,'__iter__') and not isinstance(x, (str, bytes)):
            #Python2写法
            # for sub_x in flatten(x):
            # yield sub_x
            #Python3写法
            yield from flatten(x)
        else:
            yield x
from collections import Counter
def get_most_common_word(x,COL):
    word_list = x[COL]
    FF = []
    for item in word_list:
        if type(item) == str:
            FF.append(eval(item))
        else:
            FF.append(item)
    FF = list(flatten(FF))
    c = Counter(FF)
#     为了简化后来的查询操作的时间复杂度，这里将最高词频结果返回为set结构
    return set([xx[0] for xx in c.most_common(15)])
    
```

```python
most_sea_F = get_most_common_word(table_sea,'feature')
most_jk_F = get_most_common_word(table_jk,'feature')
most_ht_F = get_most_common_word(table_ht,'feature')
most_ax_F = get_most_common_word(table_ax,'feature')
most_ep_F = get_most_common_word(table_ep,'feature')
most_ca_F = get_most_common_word(table_ca,'feature')

most_sea_D = get_most_common_word(table_sea,'Desti')
most_jk_D = get_most_common_word(table_jk,'Desti')
most_ht_D = get_most_common_word(table_ht,'Desti')
most_ax_D = get_most_common_word(table_ax,'Desti')
most_ep_D = get_most_common_word(table_ep,'Desti')
most_ca_D = get_most_common_word(table_ca,'Desti')
```

```python
def cal_weight(x,COL,MOST):
    total = len(x)
    dct = {item:0 for item in MOST}
    word_list = x[COL].values
    for item in word_list:
        if type(item) == str:
            le = eval(item)
            for item2 in le:
                if item2 in MOST:
                    dct[item2] +=1/total
        else:
            continue
    return dct         
```

```python
# 权重计算示例
cal_weight(table_sea,'feature',most_sea_F)
```

```python
# 每个目的地地区计算相应的高频词权重词典：
dct_sea_F = cal_weight(table_sea,'feature',most_sea_F)
dct_jk_F = cal_weight(table_jk,'feature',most_jk_F)
dct_ht_F = cal_weight(table_ht,'feature',most_ht_F)
dct_ax_F = cal_weight(table_ax,'feature',most_ax_F)
dct_ca_F = cal_weight(table_ca,'feature',most_ca_F)
dct_ep_F = cal_weight(table_ep,'feature',most_ep_F)

dct_sea_D = cal_weight(table_sea,'Desti',most_sea_D)
dct_jk_D = cal_weight(table_jk,'Desti',most_jk_D)
dct_ht_D = cal_weight(table_ht,'Desti',most_ht_D)
dct_ax_D = cal_weight(table_ax,'Desti',most_ax_D)
dct_ca_D = cal_weight(table_ca,'Desti',most_ca_D)
dct_ep_D = cal_weight(table_ep,'Desti',most_ep_D)
```

```python
df['VF'] = 0
df['DF'] = 0
def CAL(x,COL,DCT):
    word_list = x[COL].values
    V = []
    for item in word_list:
        if type(item) == str:
            le = eval(item)
            val = 0
            for item2 in le:
                if item2 in DCT:
                     val+=DCT[item2] 
            V.append(val)
        else:
            V.append(0)
    return V
```

```python
table_sea['VF'] = CAL(table_sea,'feature',dct_sea_F)
table_jk['VF'] = CAL(table_jk,'feature',dct_jk_F)
table_ht['VF'] = CAL(table_ht,'feature',dct_ht_F)
table_ax['VF'] = CAL(table_ax,'feature',dct_ax_F)
table_ca['VF'] = CAL(table_ca,'feature',dct_ca_F)
table_ep['VF'] = CAL(table_ep,'feature',dct_ep_F)

table_sea['DF'] = CAL(table_sea,'Desti',dct_sea_D)
table_jk['DF'] = CAL(table_jk,'Desti',dct_jk_D)
table_ht['DF'] = CAL(table_ht,'Desti',dct_ht_D)
table_ax['DF'] = CAL(table_ax,'Desti',dct_ax_D)
table_ca['DF'] = CAL(table_ca,'Desti',dct_ca_D)
table_ep['DF'] = CAL(table_ep,'Desti',dct_ep_D)
```
