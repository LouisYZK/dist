## Ch1 概率思维--贝叶斯思维指南
- 统计学是以建模为核心的
- 贝叶斯定理与统计推断
  $$P(H|D) = \frac{P(D|H)P(H)}{P(D)}$$
  $$P(H|D) \propto P(D|H)P(H)$$
  贝叶斯定理是在给定数据的情况下寻找参数的分布，所以把H理解为模型；
  - **P(H): 先验**，反应我们在观测到数据之前对参数的理解；
  - **P(D|H): 似然**， 反应给定参数下某组观测数据的可信度；
  - **P(H|D): 后验**， 贝叶斯分析的结果，反应参数的概率分布
  - **P(D): 证据**

### 抛硬币模型
- 通用模型： $p(\theta|y) \propto p(y|\theta)p(\theta)$
- 选择似然：典型的二项分布： $$p(y|\theta) = \frac{N!}{y!(N-y)!}\theta^y(1-\theta)^{(N-y)}$$
- 选择先验： 选用贝叶斯分布中最常见的**beta分布**:$$p(\theta) = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\theta^{\alpha-1}(1-\theta)^{\beta-1}$$
    选择beta分布是因为他是二项分布的共轭先验，似然的共轭先验是说乘法之后得到的后验的形式依然与先验一样；
- 计算后验：$$p(\theta|y) \propto \theta^{\alpha-1+y}(1-\theta)^{\beta-1+N-y}$$
    $$p(\theta|y) = Beta(\alpha_{prior} + y, \beta_{prior}+N-y)$$

```python
# 计算后验并画图
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

plt.style.use('seaborn')
theta_real = 0.35
trails = [0, 1, 2, 3, 4, 8, 16, 32, 50, 150]
data = [0, 1, 1, 1, 1, 4, 6, 9, 13, 48]

beta_params = [(1, 1), (0.5, 0.5), (20, 20)]
x = np.linspace(0, 1, 100)

for ind, N in enumerate(trails):
    if ind == 0:
        plt.subplot(4, 3, 2)
    else:
        plt.subplot(4, 3, ind+3)
    y = data[ind]
    for (a_prior, b_prior), c in zip(beta_params, ['b', 'g', 'r']):
#         计算beta分布的概率密度函数
        p_theta_given_y = stats.beta.pdf(x, a_prior+y, b_prior+N-y)
        plt.plot(x, p_theta_given_y, c)
        plt.fill_between(x, 0, p_theta_given_y, color=c, alpha=0.6)
    
    plt.axvline(theta_real, ymax=0.3, color='k')
    plt.plot(0, 0, label=f"experienments:{N}\n heads:{y}")
    plt.xlim(0, 1)
    plt.ylim(0, 12)
    plt.legend()

plt.tight_layout()
plt.show()
```
